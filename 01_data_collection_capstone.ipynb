```python
# Removing Column RT which is the Record Type. All of the records that have been pulled are recorded with a P for Personal Record. Because the full column lists P, it is no longer needed.
df = df.drop(columns=['RT'])
df.head()
```

```python
# Choosing to keep column SOCP for now even though the dtype is listed as an Object. 
# The dtype is Object because the main heading of each occupation starts with a four digit number and ends with XX. The XX is a sub-label.
# For example, MGR - Chief Executive and Legislators is 1110XX, while MGR-General And Operations Managers is 111021.
# More information can be found here: https://www2.census.gov/programs-surveys/acs/tech_docs/pums/data_dict/PUMS_Data_Dictionary_2016-2020.csv 
# Standard Occupational Classification 
print(df['SOCP'].unique())
```
