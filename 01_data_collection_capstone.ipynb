```python
# Removing Column RT which is the Record Type. All of the records that have been pulled are recorded with a P for Personal Record. Because the full column lists P, it is no longer needed.
df = df.drop(columns=['RT'])
df.head()
```

```python
# Removing the other object Column SOCP because it is the non-coded version of the OCCP column, whereas the OCCP column is a recode and would be better to use for this direct set of non-aggregated data. 
# SOCP = Standard Occupational Classification (SOC) codes for 2018 and later based on 2018 SOC codes
# OCCP = Occupation recode for 2018 and later based on 2018 OCC codes
# More information can be found here: https://www2.census.gov/programs-surveys/acs/tech_docs/pums/data_dict/PUMS_Data_Dictionary_2016-2020.csv 

df = df.drop(columns=['SOCP'])
df.head()
```
